## Câu 1: Nhập tuổi theo năm, in ra tuổi theo ngày
Giả sử 1 năm = 365 ngày

```python
age = int(input("Enter your age: "))
days = age * 365
print("You are", days, "days old")
```

## Câu 2: Hiển thị ngày giờ hiện tại

```python
from datetime import datetime

now = datetime.now()
print("Current date and time")
print(now.strftime("%A, %d/%m/%Y, %H:%M:%S"))
```

## Câu 3: Nhập dãy số nguyên cách nhau bởi dấu phẩy, xuất số phần tử và tổng

```python
chuoi = input("Nhap day so nguyen cach nhau boi dau phay: ")
ds = list(map(int, chuoi.split(',')))
print("So phan tu:", len(ds))
print("Tong cac phan tu:", sum(ds))
```

## Câu 4: Sinh 5 số thực ngẫu nhiên trong [0,1), tìm trung bình, min, max

```python
import random

ds = [random.random() for _ in range(5)]
print("5 so thuc ngau nhien:", ds)
print("Trung binh:", sum(ds) / len(ds))
print("Gia tri nho nhat:", min(ds))
print("Gia tri lon nhat:", max(ds))
```
